***To build a machine learning model that can predict home prices based on square feet area***

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
#loding the dataset .
df=pd.read_csv("/content/drive/MyDrive/homeprices.csv")
df.head(5)
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='*')
```

```python
df1=df.drop('price',axis='columns')
df1
```

```python
price=df.price
price
```

```python
reg = linear_model.LinearRegression()
reg.fit(df1,price)
```

```python
reg.predict([[3300]])
```

We know for the intersept


Y=M*X+B

```python
#To find the coefficients that is the value of m
m= reg.coef_
m
```

```python
b= reg.intercept_
b
```

```python
y=m*df1+b
y
```

```python
#for 3300 sqft area the y predict is
y=m*3300+b
y
```

```python
#for 5500 sqft area the y predict is
reg.predict([[5500]])
```

Importing CSV files with list of home price prediction values

```python
p = reg.predict(area)
p
```

```python
area['prices'] = p
area
```
